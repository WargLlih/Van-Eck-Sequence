```python
import pandas as pd

def eck(k,lista=False):
    graf = {'a(0)': 0} ; cont = 1
    vaneck=[0]
    for x in range(k):
        if vaneck[-1] in vaneck[:-1]:
            vaneck.append(len(vaneck[::-1][1:][:vaneck[::-1][1:].index(vaneck[-1])+1])) 
        else: vaneck.append(0)
        graf['a('+str(cont)+')'] = vaneck[-1]
        cont+=1
    if lista: return vaneck
    else: return graf
```

```python
pd.Series(eck(50)).plot(figsize=(15,7))
```

```python
pd.Series(eck(1000)).plot(figsize=(15,7))
```

```python
pd.Series(eck(20**4)).plot(figsize=(15,7))
```
